# 1.Using Regex:

```python
amz_fkrt_regex = amz_fkrt_ds.drop(columns=['uniq_id'],inplace=False).copy()
```

```python
amz_fkrt_regex
```

```python
#Function to take input(product name)  and display the result as a table
def product_finder_re():
    prod = input("Enter product name:")
    prod = prod.strip().replace(' ','').lower()
    disp_index = []
    for row in amz_fkrt_ds.iterrows():
        
        rowf = row[1]['Product Name in Flipkart'].strip().replace(' ','').lower()
        rowA = row[1]['Product Name in Amazon'].strip().replace(' ','').lower()
        
        if re.match(r'\w{{0,}}({})\w{{0,}}'.format(prod),rowf) or re.match(r'\w{{0,}}({})\w{{0,}}'.format(prod),rowA):
            disp_index.append(row[0])
    if len(disp_index)!=0:
        display(amz_fkrt_regex.loc[disp_index])
    else:
        print('Product not found')
```

**As some of the strings are not properly typed with proper spaces, and also they are in different cases(Upper and Lower) the algorithm follows the following steps:**
1. Any extra spaces in the front and back are stripped.
2. The spaces in between words are removed to convert the text into a single string. This hepls in generalizing the results.
3. All the characters in the strings, both input and output, are converted into lowercase characters.
4. All the products containing the random input pattern(string) are displayed as result. This is done using **regex** pattern matching. Hence, it is rule based.

```python
#Find the required product using this function:
product_finder_re()
```

**Drawbacks**
1. Cannot be easily generalized for a wide variety of data, since it requires manual pattern matching.
2. Involves no learning and is a hard coded algorithm.

**Advantages**
1. Extremely efficient and fast for small and specific datasets.
2. Produces accurate results.
3. Doesn't require a lot of preprocessing.
4. Doesn't require a lot of computational resources.
5. Doesn't require the exact product name to match the product.

# 2.Using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import string
from sklearn.decomposition import TruncatedSVD   
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Functions for preprocessing
punctuation = string.punctuation

#Function for lowercasing
def lower_casing(data):
      return np.char.lower(data)

#Function for removing punctuation
def remove_punct(data):
    for i in punctuation:
        data = np.char.replace(data, i , " ")  
    return data

#Functions for preprocessing the data
def preprocessor_tf(data):
    data = lower_casing(data)
    data = remove_punct(data)
    return data
```

**Assumptions:**
1. It should be noted that the single words arising as a result of the punctuation removal are ignored by the Algorithm itself and need not be seperately taken care of.
2. Stopwords are rare, but even if they exist, are not being removed because the dataset set is product name specific and are necessary for matching.
3. Also, any numbers present in the name provide special information about the product and hence are necessary for matching.  
4. Generic NLP preprocessing techniques like stemming and lemmatization are not being used to preserve word integrity as the datasets are product specific

```python
corpus = amz_fkrt_ds['Product Name in Flipkart'].to_list() + amz_fkrt_ds['Product Name in Amazon'].to_list()
corpus
```

```python
corpus = preprocessor_tf(corpus)
vect = TfidfVectorizer()
X = vect.fit_transform(corpus)#This returns X as a sparse matrix
```

```python
#The returned sparse matrix is decomposed into 500 columns
svdecom = TruncatedSVD(n_components=500)  
Xreduced_dim = svdecom.fit_transform(X)
```

```python
amz_fkrt_tf = amz_fkrt_ds.drop(columns=['uniq_id'],inplace=False).copy()
```

```python
#Function to take input(product name)  and display the result as a table
def product_finder_tfidf():
    prod = input("Enter product name:")
    prod = preprocessor_tf([prod])
    prod = vect.transform(prod)
    prod = svdecom.transform(prod)

    disp_index = []
    for row in amz_fkrt_tf.iterrows():
       
        namef = Xreduced_dim[0:20000]
        nameA = Xreduced_dim[20000:]        
        resf = cosine_similarity(namef[row[0]].reshape(1,-1),prod)
        resA = cosine_similarity(nameA[row[0]].reshape(1,-1),prod)
        
        if resf>=0.5 or resA>=0.5:
            disp_index.append(row[0])
    
    if len(disp_index)!=0:
        display(amz_fkrt_tf.loc[disp_index])
    else:
        print('Product not found')
```

# 3.Using Spacy Phrase Matcher 

```python
from spacy.matcher import PhraseMatcher
```

```python
#Function to take input(product name)  and display the result as a table
def product_finder_phrasematch():
    nlp = spacy.blank("en")
    matcher = PhraseMatcher(nlp.vocab)
    prod = input("Enter product name:")
    pattern = nlp(prod)
    matcher.add("Required", [pattern])
    disp_index = []
    
    for row in amz_fkrt_ds.iterrows():
        
        docf = nlp(row[1]['Product Name in Flipkart']) 
        doca = nlp(row[1]['Product Name in Amazon'])
        if matcher(docf) or matcher(doca):
            disp_index.append(row[0])

    if len(disp_index)!=0:
        display(amz_fkrt_tf.loc[disp_index])
    else:
        print('Product not found')

```

1. Requires the exact product name to match the product, with proper casing.
2. Cannot be generalized.
3. Requires the user to know the exact product name.
4. Average speed.


**Advantages**
1. No preprocessing is required.
2. Highly specific and accurate results.
3. Doesn't require a lot of computational resources.
4. Short and simple algorithm.
